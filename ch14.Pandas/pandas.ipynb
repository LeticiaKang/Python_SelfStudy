## 3.SQL 데이터베이스랑 연결하기

```python
import sqlite3
# 파일이 저장될 경로와 이름을 써서 데이터베이스 연결하기
conn = sqlite3.connect("DataBase/db.sqlite")

# 모든 행과 열을 가져오기
q = "select * from hanbit_books"
```

```python
# 쿼리 문자열과 데이터베이스 연결을 전달합니다.
sql_df = pd.read_sql(q, con=conn)[2:].reset_index(drop=True)
sql_df.head()
```

### 3-1.데이터베이스 저장
- 속성
    1. if_exists{'fail','replace','append'}
        - fail : 디폴트값. 테이블이 이미 존재하면 에러 발생
        - replace : 기존의 테이블을 버리고 새로 넣는다
        - append : 기존의 테이블에 추가로 넣는다.

```python
# 추가하고 싶은 데이터 DF화 하기
new_row = pd.DataFrame([["추가", "추가", "추가", "추가", "추가"]], columns = sql_df.columns)

# concat으로 중간에 원하는 값 넣기
sql_df = pd.concat([new_row, sql_df], ignore_index = True)
sql_df.shape
```

```python
# 기존에 있는 result_books대체해서 저장하기
sql_df.to_sql("result_books", con=conn, if_exists='replace')
```

## 5.데이터 병합

```python
t1 = pd.DataFrame({
        "id":[1, 2, 3, 4, 5],
        "left_val":["a", "b", "c", "d", "e"]
    })

t1
```

```python
t2 = pd.DataFrame({
        "id":[3, 4, 5, 6, 7],
        "right_val":["ㄱ", "ㄴ", "c", "ㄹ", "ㅁ"]
    })

t2
```

### 5-1.merge() 
-  같은 컬럼 이름 기준으로 교집합부분 추출, 두개의 데이터프레인의 정보를 모두 가져옴
- 공통된 컬럼 외 다른 컬럼명을 on에 넣으면 에러남

```python
pd.merge(t1, t2)
```

#### 5-1-1.공통col을 기준으로

```python
pd.merge(t1, t2, on="id")
```

#### 5-1-2.서로 다른 col을 기준으로

```python
pd.merge(t1, t2, left_on="left_val", right_on="right_val")
```

#### 5-1-3.(A여집합)∩B

```python
pd.merge(t1, t2, how='left')
```

#### 5-1-4.(B여집합)∩A

```python
pd.merge(t1, t2, how='right')
```

#### 5-1-5.합집합

```python
pd.merge(t1, t2, how='outer')
```

### 6-8.함수agg()

```python
dummy_data.groupby("contry").agg(
    {
        # 적용할 컬럼 이름
        "level":
        # 적용시킬 함수
        [   ("합계", np.sum), 
            ("평균", np.sum) 
        ]
    }
)[3:9]
```

```python
by_contry.agg(
    {
        "level":
        [np.size, np.median]
    }
)[3:9]
dummy_data.info()
```

```python
def sum_of_pow(s:pd.core.series.Series):
    # 각 그룹의 해당 열 Series 타입 데이터로 전달받음
    return sum([i ** 2 for i in s])

by_contry.agg({"level":[sum_of_pow]}).head()
```

### 6-9.함수apply()

```python
def get_diff_date(df : pd.core.frame.DataFrame):
    # apply()로 전달하는 함수의 파라미터는 각 그룹의 DataFrame 타입 데이터
    return df.df_date.max() - df.df_date.min()

by_contry.apply(get_diff_date)[30:39]
```
